###**▶ Numpy란?**
- Numerical Python의 줄임말
- 벡터, 행렬, 다차원 배열의 처리를 지원
- 수치 계산을 위해 구현된 기능 제공

<br>

###**▶ 넘파이 배열 (ndarray)**
- 넘파이의 핵심적인 객체는 다차원 배열이다.
- 넘파이는 ndarray(n차원 배열의 약자) 객체를 제공한다. 이때, 차원은 좌표계의 축과 같다.
- <b><u>ndarray 객체는 동일한 자료형의 항목들만 저장할 수 있다.</u></b>

<br>

###**▶ 수학 함수들 (np.xxx)**
- 기본 함수 : abs(절대값), sqrt(제곱근), square(제곱), exp(지수 e), log(로그), log10(밑이 10인 로그), sin, cos, tan
- 통계 함수 : sum(합계), mean(산술평균), std(표준편차), var(분산), max(최대값), min(최소값), median(중간값), argmin(최소값 인덱스), argmax(최대값 인덱스)
- 정렬 함수 : sort(정렬, axis)
- 난수 발생(numpy.random 모듈) : seed(시드 지정), shuffle(임의 섞기), rand(균등분포 추출), randint(최소/최대 중)

- 참고 사이트 : https://sacko.tistory.com/8
